```python
#Assignment 1:

rho_l = 0.74 # Density of liquid water g/cm3
rho_g = 0.036 # Density of vapour water g/cm3
micro_sigma_H = 38*10**-24 #cm2 Hydrogen microscopic total cross section
micro_sigma_O = 4.2*10**-24 #cm2 Oxygen microscopic total cross section
Nav = 6.022*10**23 #Avogadro number
MMh20 = 18.016 #g/mol Molar mass of water
macro_sigma_water_l = (2*micro_sigma_H+micro_sigma_O )*(Nav*rho_l)/MMh20 #Total Macroscopic cross section of liquid water
macro_sigma_water_g = (2*micro_sigma_H+micro_sigma_O )*(Nav*rho_g)/MMh20 #Total Macroscopic cross section of steam water
##################################################################################################################
#Volume mixture of 40%
xv = 0.4 #Volume mixture
rho_l_1_bar = 1 #Water density at 1 bar pressure 
rho_mix = rho_l*(1-xv)+rho_g*xv #Steam-Liquid mix density vith vapour phase at 40%
macro_sigma_mix = (2*micro_sigma_H+micro_sigma_O)*Nav*rho_mix/MMh20 #Total mix Macroscopic cross section 
macro_sigma_water_1_bar = (2*micro_sigma_H+micro_sigma_O)*Nav*rho_l_1_bar/MMh20 #Total Macroscopic cross section of liquid water at 1 bar pressure

print("Total Macroscopic cross section of liquid water is=", macro_sigma_water_l)
print("Total Macroscopic cross section of vapour water is=",macro_sigma_water_g)
print("Total Macroscopic cross section of the vapour-liquid =",macro_sigma_mix)
print("Total Macroscopic cross section of the vapour-liquid-1bar=",macro_sigma_water_1_bar)
```
